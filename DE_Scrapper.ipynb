```
def getCoord(address, agent):
    
    '''Collect an address and define an agent
    Return the longitude and latitute'''

    geolocator = Nominatim(user_agent=agent)
    location = geolocator.geocode(address)
    return location.latitude,  location.longitude
```

```
add = 'Hamburg'

latitude, longitude = getCoord(add,'SOG')

# create map of South Africa using latitude and longitude values
map_CT = folium.Map(location=[latitude, longitude], zoom_start=1)

# add markers to map
for lat, lng, prov in zip(df_surbs['Latitude'], df_surbs['Longitude'], df_surbs['Suburbs']):
    label = prov
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=2,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_CT)  
    
map_CT
```
